````python
out1 = kmean.kMean(blob1,verbouse=2)
out1.showKmean()
out2 = kmean.kMean(blob2,verbouse=2)
out2.showKmean()
out3 = kmean.kMean(moons,verbouse=2)
out3.showKmean()
out4 = kmean.kMean(ring,verbouse=2)
out4.showKmean()
blob16 = kmean.genBlob([(-6,6),(6,6),(6,-6),(-6,-6),(-3,0),(0,0),(3,0),(0,-4),(-5,12),(0,12),(5,12),(0,-12)],200)
blob16.show()
out5 = kmean.kMean(blob16,verbouse=2)
out5.showKmean()

````

````python
from sklearn.metrics import silhouette_score

sil = []
kmax = 10

# dissimilarity would not be defined for a single cluster, thus, minimum number of clusters should be 2
for k in range(2, kmax+1):
    out1 = kmean.kMean(blob2,k)
    out2 = kmean.kMean(blob2,k)
    if kmean.getMeanError(out1) < kmean.getMeanError(out2):
        sil.append(silhouette_score(out1.xy, out1.center, metric = 'euclidean'))
    else:
        sil.append(silhouette_score(out2.xy, out2.center, metric = 'euclidean'))
    #out2.showKmean()
print(sil)
from matplotlib import pyplot as plt
plt.plot(sil)
````
